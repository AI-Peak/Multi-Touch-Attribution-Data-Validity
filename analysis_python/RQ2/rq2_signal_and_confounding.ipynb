## RQ2 Conclusion

Channel identity is weak at row level: the chi-square p-value is high and row-channel AUC is near random. User-level prediction improves mainly because journey length captures repeated exposure. Markov removal effects should therefore be treated as diagnostic output, not causal channel credit.
